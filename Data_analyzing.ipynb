```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_excel(r'H:\My Drive\Umich\Winter 2022\OM 660\Term Project\unemplyment_rate_all_states_2017_to_2021.xlsx')
df
```

```python
Alabama = df[df['State'] == 'Alabama'] 
Arizona = df[df['State'] == 'Arizona']
```

```python
plt.figure(1, figsize=(20,15))
plt.subplot(211)
plt.plot(Alabama['Date'], Alabama['Rate'], color='red', marker='o', label="Alabama")
plt.plot(Arizona['Date'], Arizona['Rate'], color='blue', marker='o', label="Arizona")
plt.legend(loc="upper left")
plt.title('Alabama Unemployment Rate Vs Year', fontsize=14)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Unemployment Rate', fontsize=14)
plt.grid(True)
plt.show()
```
