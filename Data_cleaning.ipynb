```python
# source: https://beta.bls.gov/dataQuery/find?st=400&r=100&q=unemployment&fq=cg:[Geography]&fq=cc:[States+and+Territories]&more=0
#df = pd.read_excel(r'H:\My Drive\Umich\Winter 2022\OM 660\Term Project\DataFinder-20220205194712.xlsx', index_col=None, header=None)
#print (df)

# get the excel file from PC
xls = pd.ExcelFile(r'H:\My Drive\Umich\Winter 2022\OM 660\Term Project\DataFinder-20220205194712.xlsx')
```

```python
len(xls.sheet_names)

#append names of all sheets into a list
sheets = []
sheets = xls.sheet_names
```

```python
len(sheets)
sheets
```

```python
#open a specific sheet in excel
df = pd.read_excel(xls, sheet_name=sheets[0])
df
```

```python
#find and append names of states from excel sheets to a list
states = []

i = 0
for sheet in sheets:
    states.append(pd.read_excel(xls, sheet_name=sheets[i])['Unnamed: 1'][6])
    i = i + 1
print(states)

```

```python
#find all duplicates sheets by states
states_dup = []
for state in set([x for x in states if states.count(x) > 1]):
    states_dup.append(state)
states_dup
```

```python
#find location of duplicated sheets
states_dup_sheets = []
i = 0
for state in states_dup:
    states_dup_sheets.append(states.index(states_dup[i]))
    i = i +1
states_dup_sheets
```

```python
#open duplicates sheets to find out the issues
df = pd.read_excel(xls, sheet_name=states_dup_sheets[0])
df1 = pd.read_excel(xls, sheet_name=states_dup_sheets[0]+1)
```

```python
#find and append measure data type from excel sheets to a list
data_type = []

i = 0
for sheet in sheets:
    data_type.append(pd.read_excel(xls, sheet_name=sheets[i])['Unnamed: 1'][5])
    i = i + 1
print(data_type)
```

```python
# find location of unemployment data
indices = [i for i, x in enumerate(data_type) if x == "unemployment"]
indices
```

```python
# get names of all the wrong sheets
sheets_to_del = []

i = 0
for index in indices:
    sheets_to_del.append(sheets[indices[i]])
    i = i + 1

sheets_to_del
```

```python
#delete wrong sheets
workbook=openpyxl.load_workbook(xls)

i = 0
for sheet in sheets_to_del:
    std=workbook.get_sheet_by_name(sheets_to_del[i])
    workbook.remove_sheet(std)
    i = i + 1

workbook.save(r'H:\My Drive\Umich\Winter 2022\OM 660\Term Project\unemplyment_rate_by_states.xlsx')
```
