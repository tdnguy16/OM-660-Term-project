```python
import pandas as pd

# get the excel file from PC
xls1 = pd.ExcelFile(r'H:\My Drive\Umich\Winter 2022\OM 660\Term Project\unemplyment_rate_by_states.xlsx')

#append names of all sheets into a list
sheets = []
sheets = xls1.sheet_names

#find and append names of states from excel sheets to a list
states = []
states_names = states

i = 0
for sheet in sheets:
    states.append(pd.read_excel(xls1, sheet_name=sheets[i])['Unnamed: 1'][6])
    i = i + 1

# convert list of states to string
states = [str(r) for r in states]

print(states)

#find all duplicates sheets by states
states_dup = []
for state in set([x for x in states if states.count(x) > 1]):
    states_dup.append(state)
states_dup

#open a specific sheet in excel
df = pd.read_excel(xls1, sheet_name=sheets[1])
df

# get the shape of all sheets to make sure they are identitcal
for sheet in sheets:    
    df = pd.read_excel(xls1, sheet_name=sheet)
    print(df.shape)

# Assign sheets to state name in list of states
i = 0
for sheet in sheets:
    states[i] = df = pd.read_excel(xls1, sheet_name=sheets[i])
    i = i + 1

# delete blank unnecessary rows and columns of each sheets
for index, state in enumerate(states):
    state.drop(state.columns[4:], inplace=True, axis=1)
    state.drop(state.index[:11], inplace=True, axis=0)
    #state.drop(state.iloc[0], inplace=True, axis=0)
    states[index] = state

# create a dictionary of states and umemployment data
unemployment_rate_dict = {}
for name in states_names:
    for state in states:
        unemployment_rate_dict[name] = state
        states.remove(state)
        break

#Create new column of state name in each sheet
for state in states_names:
    unemployment_rate_dict.get(state)['State'] = state

#concatenate all sheets into 1 dataframe
all_state_data = pd.concat(unemployment_rate_dict.values(), ignore_index=True)
print(all_state_data)

print(all_state_data.columns)

#rename dataframe columns
all_state_data.rename(columns={'Local Area Unemployment Statistics': 'Year', 'Unnamed: 1': 'Month', 'Unnamed: 2': 'Date', 'Unnamed: 3': 'Rate'}, inplace=True)

# save dataframe as an excel file
all_state_data.to_excel(r'H:\My Drive\Umich\Winter 2022\OM 660\Term Project\unemplyment_rate_all_states.xlsx') 
```
